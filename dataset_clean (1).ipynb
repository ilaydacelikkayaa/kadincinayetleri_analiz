```python
df['Tarih:']=pd.to_datetime(df['Tarih:'], format="%d/%m/%Y",errors="coerce")
```

```python
df['Maktülün yaşı:'].isna().sum()
```

```python
df["Maktülün yaşı:"] = df["Maktülün yaşı:"].fillna("Bilinmiyor")
```

```python
df['İl/ilçe:']=df['İl/ilçe:'].fillna('Bilinmiyor')
```

```python
df["Öldürülme şekli:"] = df["Öldürülme şekli:"].fillna("Bilinmiyor")
df["Failin durumu:"] = df["Failin durumu:"].fillna("Bilinmiyor")
df["Korunma talebi:"] = df["Korunma talebi:"].fillna("Bilinmiyor")
df["Kim tarafından öldürüldü:"] = df["Kim tarafından öldürüldü:"].fillna("Bilinmiyor")
```

```python
korunma_sayilari = df["Korunma talebi:"].value_counts()
top3 = korunma_sayilari.head(3)
diger = korunma_sayilari[3:].sum()

korunma_grafik = pd.concat([top3, pd.Series({"Diğer": diger})])

plt.figure(figsize=(4,4))
plt.pie(korunma_grafik, labels=korunma_grafik.index,
        autopct="%1.1f%%", startangle=90)
plt.title("Kadınların Korunma Talebi (Gruplandırılmış)")
plt.axis("equal")
plt.show()
```

```python
df["Korunma Talebi"] = df["Korunma talebi:"].str.strip().str.lower()
df["Fail Tutuklandı mı?"] = df["Failin durumu:"].str.strip().str.lower()


korunanlar = df[df["Korunma Talebi"] == "var"]

tutuklu_durum = korunanlar["Fail Tutuklandı mı?"].value_counts()

plt.figure(figsize=(8,5))
tutuklu_durum.plot(kind="bar", color="indianred", edgecolor="black")
plt.title("Korunma Talep Eden Kadınların Faillerinin Tutuklanma Durumu")
plt.xlabel("Tutuklanma Durumu")
plt.ylabel("Kişi Sayısı")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

Bu durum, hem adli sistemdeki şeffaflık eksikliğini hem de koruyucu

```python
df["ay"]=df['Tarih:'].dt.month
```

```python
aylik_sayilar=df["ay"].value_counts().sort_index()
plt.figure(figsize=(10,6))
aylik_sayilar.plot(kind="barh",color="darkslateblue")
plt.show()
```

Grafikte 8. ve 9. aylar (yani Ağustos ve Eylül) diğer aylara göre en yüksek cinayet sayılarına sahip.

Bu durum yaz aylarının sonuna doğru kadın cinayetlerinde bir artış yaşandığını göstermektedir. Tatil dönemleri, sosyal ilişkilerde artış gibi etkenler bu artışta rol oynayabilir.
En düşük cinayet sayısı:Ocak ve Kasım

Bu aylar genelde kış aylarına denk geldiği için dışarıda sosyalleşmenin azaldığı, insanlar arası etkileşimin (ve belki çatışmanın) azaldığı dönemler olabilir.

```python

akraba_failler = df[df["Kim tarafından öldürüldü:"].str.lower().str.contains("akraba", na=False)]

akraba_iller = akraba_failler["İl/ilçe:"].value_counts()

akraba_iller_filtered = akraba_iller.drop("Bilinmiyor", errors="ignore")

akraba_iller_filtered = akraba_iller_filtered.sort_values(ascending=True)

plt.figure(figsize=(10,6))
akraba_iller_filtered.plot(kind="barh", color="darkred", edgecolor="black")
plt.title("Akraba Tarafından Öldürülen Kadınlar (İllere Göre)")
plt.xlabel("Olay Sayısı")
plt.ylabel("İl")
plt.tight_layout()
plt.show()
```

Verilere göre “akraba” failli kadın cinayetleri Türkiye'nin birçok ilinde görülmekle birlikte, en çok İstanbul ve Güneydoğu illerinde öne çıkmaktadır.
Bu durum hem büyükşehirlerdeki geleneksel toplum yapısında ataerkil baskının ve törelerin etkisini göstermektedir.
